```python
import pandas as pd
```

```python
df = pd.read_csv("Source/2020_raw.csv")
```

```python
df
```

```python
# new data frame with split value columns 
new = df["Full Name"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df["First Name"]= new[1] 
  
# making separate last name column from new data frame 
df["Last Name"]= new[0] 
```

```python
filename = input("What would you like to name this new csv file?")
df.to_csv(filename+"_split.csv")
```
